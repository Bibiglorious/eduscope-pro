#### Importing files and loading dataset

```python
import pandas as pd
print("Pandas imported successfully!")
```

```python
# Load the engineered data
df = pd.read_csv('../data/processed/engineered_data.csv')
```

#### Create a table for districts

```python
# Create district_id columnn
df['district_id'] = df['student_id'].astype(str).str[:3]
```

```python
# Create district_name columnn
# Using if-else statements to asign district names
def assign_district_name(student_id):
    if student_id == '001':
        return 'Washington District'
    elif student_id == '002':
        return 'New York District'
    else:
        return 'Unknown District'

```

```python
# Apply the function to create district_name column
df['district_name'] = df['student_id'].apply(assign_district_name)
```

#### Calculate district Insights

```python
# Avergae GPA per district
average_gpa_per_district = df.groupby(['district_id','district_name'])['gpa'].mean().reset_index()
```

```python
average_gpa_per_district
```

```python
# Number of at-risk students per district 
at_risk_students_per_district = df[df['risk_score'] == True].groupby(['district_id', 'district_name']).size().reset_index(name='at_risk_count')
```

```python
at_risk_students_per_district
```

```python
# Engagement rate per district
engagement_rate_per_district = df.groupby(['district_id', 'district_name', 'engagement_level']).size().unstack(fill_value=0)
engagement_rate_per_district ['total_students'] = engagement_rate_per_district .sum(axis=1)
engagement_rate_per_district ['high_engagement_rate'] = (engagement_rate_per_district ['High'] / engagement_rate_per_district ['total_students'] * 100).round(2)
```
